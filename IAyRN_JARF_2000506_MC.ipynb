```python
# Exámen Medio Curso
# Nombre - Matrícula: José Alberto Romero Francisco-2000506
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 1. Cargar el datasheet con pandas
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df = pd.read_csv(url)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 2. Imprime las últimas 5 filas del dataset
print("Últimas 5 filas del dataset:")
display(df.tail())

# Genera un resumen estadístico descriptivo de las columnas numéricas
print("\nResumen estadístico descriptivo:")
display(df.describe())
```

```python
# Identifica las filas donde x, y o z son 0
rows_with_zero_dimensions = df[(df['x'] == 0) | (df['y'] == 0) | (df['z'] == 0)]

print(f"Número de filas con x, y o z igual a 0 antes de eliminar: {len(rows_with_zero_dimensions)}")

# Elimina esas filas del DataFrame
df_cleaned = df[~((df['x'] == 0) | (df['y'] == 0) | (df['z'] == 0))].copy()

print(f"Número de filas después de eliminar: {len(df_cleaned)}")
print("\nPrimeras 5 filas del DataFrame después de eliminar filas con dimensiones cero:")
display(df_cleaned.head())
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df_cleaned['price'])
plt.title('Distribución del Precio de los Diamantes')
plt.ylabel('Precio')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(df_cleaned['volumen_aprox'], df_cleaned['price'], alpha=0.5)
plt.title('Relación entre Volumen Aproximado y Precio del Diamante')
plt.xlabel('Volumen Aproximado')
plt.ylabel('Precio')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
cut_counts = df_cleaned['cut'].value_counts()

plt.figure(figsize=(10, 6))
cut_counts.plot(kind='bar', color='skyblue')
plt.title('Cantidad de Diamantes por Tipo de Corte')
plt.xlabel('Tipo de Corte')
plt.ylabel('Cantidad de Diamantes')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Crea la nueva columna 'volumen_aprox'
df_cleaned['volumen_aprox'] = df_cleaned['x'] * df_cleaned['y'] * df_cleaned['z']

print("\nPrimeras 5 filas del DataFrame con la nueva columna 'volumen_aprox':")
display(df_cleaned.head())
```
